```python
import numpy as np
import matplotlib .pyplot as plt
```

```python


plt.rcParams["figure.figsize"] = (8, 8)
for i in range(5):  
  binomial_dist= np.random.binomial(100,.4,200*10**i) # picks out 200*10^i samples from a binomial distribution with n=100 & p=.4
  plt.hist(binomial_dist, bins = 100)
  plt.savefig('Sample size {}'.format(200*10**i))
  plt.show()

```

```python
binomial_dist = np.random.binomial(100,.4,10000)

print('The theoretical avearge is: {}, the experimentally calculated average is: {}'.format(40,np.average(binomial_dist)))

print('The theoretical deviation is: {}, the experimentally calculated deviation is: {}'.format((100*.4*.6)**.5,np.std(binomial_dist)))
```
